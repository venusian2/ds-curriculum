# Confidence Interval

Objectives:

One way to get many of the advantages of hypothesis the testing without the stark black/white, reject/fail-to-reject dichotomies of hypothesis testing is to use **confidence intervals**. A confidence interval is defined as a range of values that's likely to include a population value with a certain degree of confidence. It's often expressed as a percentage whereby a population mean lies between an upper and lower interval. 

For example, when we state that we have a 95% confidence level, it means that, should you repeat an experiment or survey over a over again, 95 percent of the time your results will match the results you get from a population.


## Confidence Interval for Different Types of Distributions

Generally speaking, a confidence interval can be applied in different types of hypothesis testing setup. In this section, we mainly focus on explaining how a confidence interval is calculated and applied in sampling distribution. 

1. Sample Means Distribution, where population standard deviation is known or given.

2. Sample Means Distribution, where population standard deviation is unknown or not given.


### Sample Means Distribution (Population SD is known)

In inferential statistics, we are trying to use characteristics of a sample (i.e. a statistic) to estimate the characteristics of the population (i.e. a parameter). When we obtain a random sample and calculate a sample statistic from that sample, the sample statistic is a random variable (each sample got selected has different statistics). The population parameters, however, are fixed. If the statistic is a random variable, we can then find the distribution, mean, and standard deviation.

The **sample mean distribution / sampling distribution** of a statistic is a probability distribution based on a large number of samples of size *n* from a given population. Once we calculate the sampling distribution, we can apply the confidence interval for statistical inference.

When the population parameter (standard deviation) is given in a problem, we can calculate the confidence interval with the following steps.

Approximately normal and following Normal Distribution $N(\mu, \sigma)$

Unbiased Estimated Mean: 

$$\bar{x} = \frac{\sum(x_i)}{n}$$

Unbiased Estimated Standard Error: 

$$\sigma_{\bar{x}} \approx \frac{\sigma}{\sqrt{n}}$$

**Confidence Interval:**

left endpt.: $\bar{x} - z\times\frac{\sigma}{\sqrt{n}}$

right endpt.: $\bar{x} + z\times\frac{\sigma}{\sqrt{n}}$


### Sample Means Distribution (Population SD is unknown)

When the population standard deviation is not known, we have to approximate the sample mean's distribution with the following steps.

Approximately normal and following Student's t distribution

Unbiased Estimated Mean: 

$$\bar{x} = \frac{\sum(x_i)}{n}$$

Unbiased Estimated Standard Error: 

$$\sigma_{\bar{x}} \approx \frac{s}{\sqrt{n}}$$

**Confidence Interval:**

left endpt.: $\bar{x} - t\times\frac{s}{\sqrt{n}}$

right endpt.: $\bar{x} + t\times\frac{s}{\sqrt{n}}$


## Bootstrap

Re-sampling methods are usually used to create new samples for statistical inference. **Bootstrapping** is sampling with replacement from observed data to estimate the variability in a statistic of interest. A common application of the bootstrap is to assess the accuracy of an estimate based on a sample of data from a large population.
